```python
df.drop(columns=['id',"Unnamed: 32"],inplace=True)
```

# label encoding for target feature

```python
df['diagnosis']=df['diagnosis'].replace({"M":1,"B":0})
```

```python
df
```

# Segregate x and y

```python
x=df.drop(columns=["diagnosis"])
y=df['diagnosis']
```

```python
x
```

```python
y
```

# Scaling

```python
ss= StandardScaler()
```

```python
x_scaled=ss.fit_transform(x)
x_scaled
```

# splitting x and y in train and test 

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
print("The shape of x_train",x_train.shape)
print("The shape of y_train",y_train.shape)
print("The shape of x_test",x_test.shape)
print("The shape of y_test",y_test.shape)
```

```python
Lr= LogisticRegression()
```

```python
Lr.fit(x_train,y_train)
```

```python
y_pred=Lr.predict(x_test)
```

```python
y_pred_train=Lr.predict(x_train)
```

```python
accuracy_score(y_pred,y_test)
```

```python
accuracy_score(y_pred_train,y_train)
```

```python
df['diagnosis'].value_counts() # data is not imbalanced
```

```python
cm=confusion_matrix(y_pred,y_test)
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(data=cm, linewidth=.5, annot=True, fmt="g", cmap="Set1")
```

```python
truepov=cm[0,0]
truenev=cm[1,1]
falsepov=cm[1,0]
falseneg=cm[0,1]
```

```python
recall = (truepov)/(truepov+falseneg)
```
